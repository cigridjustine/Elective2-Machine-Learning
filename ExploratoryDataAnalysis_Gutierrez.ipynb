## Importing libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns                       
import matplotlib.pyplot as plt 

from google.colab import files
data_to_load = files.upload()

%matplotlib inline     
sns.set(color_codes=True)
```

## Loading the dataframe

```python
import io
df = pd.read_csv(io.BytesIO(data_to_load['Housing.csv']))
```

## Checking the data types

## Dropping irrelevant features

```python
df = df.drop(['airconditioning', 'hotwaterheating'], axis=1)
df.head(10)
```

## Renaming columns

```python
df = df.rename(columns={"bedrooms": "BR", "stories": "STR", "mainroad": "MR"})
df.head(10)
```

```python
df.shape
```

## Dropping duplicated rows

```python
duplicate_rows_df = df[df.duplicated()]
print("Duplicated rows: ", duplicate_rows_df.shape)
```

```python
df.count()  
```

```python
df = df.drop_duplicates()
df.head(10)
```

```python
df.count()
```

## Dropping missing values

```python
print(df.isnull().sum())
```

```python
df = df.dropna()   
df.count()
```

```python
print(df.isnull().sum()) 
```

# Detecting Outliers

## Boxplot

```python
sns.boxplot(x=df['BR'])
```

```python
sns.boxplot(x=df['bathrooms'])
```

```python
sns.boxplot(x=df['STR'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.shape
```

## Scatterplot

```python
sns.scatterplot(x="BR", y="price", palette="rocket", data = df)
```
