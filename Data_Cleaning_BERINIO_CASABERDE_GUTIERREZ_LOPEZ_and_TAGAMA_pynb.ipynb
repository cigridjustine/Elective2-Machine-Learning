Problem 2: In `Email Address` we can see that there is a wrong input and a missing dot. So lets remove the wrong input and replace it with 'NA', and for the missing dot lets remove the space and replace it with dot

```python
data['Email Address'].unique()
```

```python
data['Email Address'] = data['Email Address'].str.replace("Emmanuel P. Larion","NA")#remove the wrong input and replace with NA
data['Email Address'] = data['Email Address'].str.replace("m ","m")#remove the space at the end of the email address
data['Email Address'] = data['Email Address'].str.replace(" ",".")#remove the space and replace with dot

data['Email Address'].unique()
#so now the `Email Address` is now clean
```

```python
NewYearRes = data[[ "Profession","Age", "Gender", "Resolution"]]
display(NewYearRes)
```

```python
NewYearRes.sort_values("Profession")
```

```python
def percent(plot, feature):
    total = len(feature)
    for p in plot.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.15
        y = p.get_y() + p.get_height()
        plot.annotate(percentage, (x, y), size = 12)
    plt.show

fig, ax = plt.subplots(figsize=(20, 6))
profession_plot = sns.countplot(x = 'Profession', data = NewYearRes)
plt.title('Designation/Profession of Respondents')

percent(profession_plot, NewYearRes.Profession)
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
age_plot = sns.distplot(NewYearRes['Age'], bins = 20)
plt.title('Age of Respondents')
```

```python
fig, ax =plt.subplots(figsize=(15, 6))
sex_plot = sns.countplot(x = 'Gender', data = NewYearRes)
plt.title('Gender of Respondents')
percent(sex_plot, NewYearRes.Gender)
```
